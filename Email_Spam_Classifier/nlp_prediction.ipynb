```python
import pandas as pd
```

```python
df = pd.read_csv('SMSSpamCollection.csv', sep='\t',header=None,names=["labels","messages"])
```

```python
df.head()
```

```python
df["messages"][0]
```

```python
import nltk
import re
from nltk.corpus import stopwords
from nltk.stem.porter import PorterStemmer
nltk.download("stopwords")
```

```python
df["messages"][0]
```

```python
words = re.sub('[^a-zA-Z]',' ',df["messages"][0]).lower().split()
```

```python
ps = PorterStemmer()
```

```python
" ".join([ps.stem(word )for word in words if word not in stopwords.words('english')])
```

```python

corpus = []
for i in range(0,len(df)):
    review = re.sub('[^a-zA-Z]',' ',df["messages"][i])
    review = review.lower()
    review = review.split()
    review = [ps.stem(word) for word in review if review not in stopwords.words('english')]
    review = " ".join(review)
    corpus.append(review)
```

```python
corpus
```

```python
from sklearn.feature_extraction.text import CountVectorizer
```

```python
cv = CountVectorizer(max_features=2500)
```

```python
X = cv.fit_transform(corpus).toarray()
```

```python
X.shape
```

```python
df.labels
```

```python
y = pd.get_dummies(df['labels'],drop_first=True)
```

```python
from sklearn.model_selection import train_test_split
```

```python
X_train,X_test,y_train,y_test= train_test_split(X,y,test_size=0.25,random_state=15)
```

```python
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import accuracy_score
model = MultinomialNB()
model.fit(X_train,y_train)
print(model.score(X_train,y_train))
y_pred = model.predict(X_test)
print(accuracy_score(y_test,y_pred))
```
